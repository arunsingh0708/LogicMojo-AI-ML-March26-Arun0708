Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute:
Mean
Median
Standard deviation
Variance (for each feature)
3. Identify:
Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
df_raw = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df = df_raw.copy()

# Task 1: Extract feature matrix as NumPy array
X = df.values
print("Shape of feature matrix:", X.shape)

# Task 2: Compute mean, median, standard deviation, and variance for each feature
mean = df.mean()
median = df.median()
std_dev = df.std()
variance = df.var()
print("\nMean:\n", mean)
print("\nMedian:\n", median)
print("\nStd dev:\n", std_dev)
print("\nVariance:\n", variance)

# Task 3: Identify the feature with highest variability
max_var_feature = variance.idxmax()
print("\nFeature with highest variance:", max_var_feature)
print("Variance value:", variance.max())

# Task 4: Reshape one feature to (n,1) — ML models expect 2D input (n_samples, n_features)
petal_length = df[['petal length (cm)']]
print("\nShape of reshaped feature:", petal_length.shape)
```

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column

Q8. Group-Based Insights (Customer Segmentation Logic)

Context: ML models often depend on aggregated patterns.

Dataset: Tips

Tasks:

1. Compute:
Average tip by day
Total bill by gender
2. Create new feature:
tip_percentage
3. Identify:
Which day has highest tipping behavior
4. Explain:
How this helps in segmentation models

```python
import seaborn as sns
import pandas as pd

df_tips = sns.load_dataset("tips")

# Task 1: Compute average tip by day and total bill by gender
avg_tip_by_day = df_tips.groupby('day')['tip'].mean()
print("Average tip by day:\n", avg_tip_by_day)

total_bill_by_gender = df_tips.groupby('sex')['total_bill'].sum()
print("\nTotal bill by gender:\n", total_bill_by_gender)

# Task 2: Create tip_percentage feature
df_tips['tip_percentage'] = (df_tips['tip'] / df_tips['total_bill']) * 100
print(df_tips[['total_bill', 'tip', 'tip_percentage']].head())

# Task 3: Identify which day has highest tipping behavior
avg_tip_pct_by_day = df_tips.groupby('day')['tip_percentage'].mean()
best_day = avg_tip_pct_by_day.idxmax()
print("\nAverage tip % by day:\n", avg_tip_pct_by_day)
print("\nDay with highest tipping behavior:", best_day)

# Task 4: Grouping reveals customer segments — models can learn patterns like
# weekend high-spenders or generous tippers for targeted recommendations
```

Q10. Data Transformation for Model Input (NumPy)

Context: ML models require structured numeric arrays.

Tasks:

1. Create two arrays of shape (3,2)
2. Perform:
Vertical stacking
Horizontal stacking
3. Reshape into (2,6)
4. Explain:
Why reshaping is required in ML pipelines

```python
import numpy as np

# Task 1: Create two arrays of shape (3,2)
A = np.array([[1, 2],
              [3, 4],
              [5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

print("Shape of A:", A.shape)
print("Shape of B:", B.shape)

# Task 2: Perform vertical and horizontal stacking
v_stack = np.vstack((A, B))
print("\nVertical Stack:\n", v_stack)
print("Shape:", v_stack.shape)

h_stack = np.hstack((A, B))
print("\nHorizontal Stack:\n", h_stack)
print("Shape:", h_stack.shape)

# Task 3: Reshape into (2,6)
reshaped = v_stack.reshape(2, 6)
print("\nReshaped Array:\n", reshaped)
print("Shape:", reshaped.shape)

# Task 4: ML models require (n_samples, n_features) format — vstack adds samples,
# hstack adds features, and reshape ensures the array fits the expected input shape
```

Q11. End-to-End ML Preprocessing Pipeline (Titanic)

Context: Prepare dataset for ML model training

Tasks:

1. Load dataset
2. Handle missing values
3. Encode categorical variables (sex, embarked)
4. Normalize numerical features
5. Convert to NumPy array

Output:

Final feature matrix ready for model training

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Task 1: Load dataset
df = sns.load_dataset("titanic")

# Task 2: Handle missing values
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df = df.drop(columns=['deck'])

# Task 3: Encode categorical variables
df['sex'] = df['sex'].map({'male': 0, 'female': 1})
df = pd.get_dummies(df, columns=['embarked'], drop_first=True)

# Task 4: Select relevant features
features = ['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked_Q', 'embarked_S']
X = df[features].copy()
y = df['survived']

# Task 5: Normalize numerical features
num_cols = ['age', 'fare']
X[num_cols] = (X[num_cols] - X[num_cols].mean()) / X[num_cols].std()

# Task 6 (Output): Convert to NumPy arrays — final feature matrix ready for model training
X_np = X.values
y_np = y.values
print("Final Feature Matrix Shape:", X_np.shape)
print("Sample Features:\n", X_np[:5])
print("Target Shape:", y_np.shape)
```

Q12. Feature Selection using Statistical Understanding

Context: Not all features improve model performance.

Dataset: Iris

Tasks:

1. Compute correlation matrix
2. Identify highly correlated features
3. Drop redundant features
4. Explain:
Impact on overfitting and model performance

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Task 1: Compute correlation matrix
corr_matrix = df.corr()
print(corr_matrix)

# Task 2: Identify highly correlated feature pairs (threshold > 0.9)
high_corr = np.where(np.abs(corr_matrix) > 0.9)
high_corr_pairs = [(corr_matrix.index[x], corr_matrix.columns[y])
                   for x, y in zip(*high_corr) if x != y]
print("\nHighly correlated pairs:\n", high_corr_pairs)

# Task 3: Drop redundant feature
df_reduced = df.drop(columns=['petal width (cm)'])
print("\nRemaining features:\n", df_reduced.columns)

# Task 4: Highly correlated features cause multicollinearity — removing them
# reduces overfitting and produces a more generalizable model
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("titanic")
df['age'] = df['age'].fillna(df['age'].median())

# Task 1 & 2: Create age_group feature using .apply() with custom function
def categorize_age(age):
    if age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df['age_group'] = df['age'].apply(categorize_age)

# Task 3: Compute survival rate per age group
survival_by_group = df.groupby('age_group')['survived'].mean()
print(survival_by_group)

# Task 4: Children have the highest survival likelihood; Seniors the lowest —
# age_group is a more interpretable feature than raw age for model training
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

# 1. Generate 1000 random error values
np.random.seed(42)
errors = np.random.randn(1000)

# 2. Compute mean and standard deviation
mean = np.mean(errors)
std = np.std(errors)
print("Mean:", round(mean, 4))
print("Standard Deviation:", round(std, 4))

# 3. Identify outliers using mean ± 2*std
lower = mean - 2 * std
upper = mean + 2 * std
outliers = errors[(errors < lower) | (errors > upper)]
print("\nOutlier threshold: [{:.4f}, {:.4f}]".format(lower, upper))
print("Number of outliers:", len(outliers))
print("Outlier values (first 10):", outliers[:10].round(4))

# 4. High variance in model performance means the model's prediction errors
# fluctuate widely across samples — it performs well on some inputs but poorly
# on others. This typically signals overfitting: the model memorized training
# data and fails to generalize to unseen examples.
```

Q15. Business + ML Insight Problem (Tips Dataset)

Context: Convert data analysis into decision-making

Tasks:

1. Create:
tip_percentage
2. Group by:
day and time
3. Identify:
Highest revenue segment
4. Provide:
Recommendation for business strategy
How ML model can use this insight

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("tips")

# Task 1: Create tip_percentage feature
df['tip_percentage'] = (df['tip'] / df['total_bill']) * 100

# Task 2: Group by day and time, compute revenue metrics
grouped = df.groupby(['day', 'time']).agg(
    avg_tip_percentage=('tip_percentage', 'mean'),
    avg_total_bill=('total_bill', 'mean'),
    count=('total_bill', 'size')
)
grouped['revenue_proxy'] = grouped['avg_total_bill'] * grouped['count']
print(grouped)

# Task 3: Identify highest revenue segment
best_segment = grouped['revenue_proxy'].idxmax()
print("\nHighest revenue segment:", best_segment)

# Task 4: Saturday Dinner drives the most revenue — allocate more staff/marketing
# on weekends; ML model can use day and time as features to predict tip behavior
```
