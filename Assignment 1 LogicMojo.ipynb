Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.

transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

Write a Python program that performs the following analysis:
1. Identify and display all transaction values that appear more than once.
2. Detect transactions greater than ₹5000 as these may require fraud monitoring.
3. Calculate the total spending for the day.
4. Create a new list containing only unique transaction values.
5. Sort the unique transaction list in descending order to view the largest transactions first.

```python
#1 Identify and display all transaction values that appear more than once.
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
duplicateAmounts = []
fraudulentTransactions = []
uniqueTransactions = []
totalSpendingInDay = 0
for i in range(len(transactions)):
    for j in range(i+1,len(transactions)):
       if transactions[i]== transactions[j] and transactions[i] not in duplicateAmounts:
         duplicateAmounts.append(transactions[i])
        
print("Transactions appearing more than once ",duplicateAmounts)

#2. Detect transactions greater than ₹5000 as these may require fraud monitoring.
for amount in transactions:
    if amount > 5000 and amount not in fraudulentTransactions:
        fraudulentTransactions.append(amount)
print("Fraudulent transactions are :",fraudulentTransactions)

#3. Calculate the total spending for the day.
for amount in transactions:
    totalSpendingInDay = totalSpendingInDay + amount
print("Total spending in day is: ",totalSpendingInDay)

#4. Create a new list containing only unique transaction values.
for amount in transactions:
    if transactions.count(amount)>=1 and amount not in uniqueTransactions:
        uniqueTransactions.append(amount)
print("Unique transaction values: ",uniqueTransactions)

#5. Sort the unique transaction list in descending order to view the largest transactions first.
uniqueTransactions.sort(reverse=True)
print ("Unique transactions sorted in descending order: ",uniqueTransactions)

    
    
```

Question 2:- A dataset exported from a database contains the following customer age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
Some entries are invalid because they are either negative numbers or missing values (None).

Write a Python program that:
1. Iterates through the dataset and removes all invalid age values.
2. Stores only valid ages in a new list.
3. Calculates the average age of valid customers.
4. Displays the list of customers whose age is greater than 30.
5. Prints how many invalid entries were removed during the cleaning process.

```python
# 1. Iterates through the dataset and removes all invalid age values.
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
invalidCount = 0
ageSum = 0
customerAboveThirty = []
for val in ages:
    if val is None or val < 0:
        invalidCount = invalidCount + 1
        ages.remove(val)

#2. Stores only valid ages in a new list
print ("Valid ages are: ",ages)

#3. Calculates the average age of valid customers.
for age in ages:
    ageSum = ageSum + age
print("Average age is ", ageSum/len(ages))

#4.Displays the list of customers whose age is greater than 30.
for age in ages:
    if age > 30:
        customerAboveThirty.append(age)
print("List of Customers above 30: ",customerAboveThirty)

#5.Prints how many invalid entries were removed during the cleaning process
print("Invalid entry count removed is: ",invalidCount)
    
    
```

Question 4:- A web platform records the hours when users log into the system.

login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
Write a Python program that:

1. Separates login hours into even-hour logins and odd-hour logins.
2. Identifies login hours that fall into peak activity time (hours greater than 5).
3. Sorts the login hours in ascending order.
4. Calculates how many logins occurred during peak hours.

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
even_login_hours = []
odd_login_hours = []
peak_activity_login_hour =[]
for hour in login_hours:
    if hour%2==0:
        even_login_hours.append(hour)
    else:
        odd_login_hours.append(hour)
print("Even login hours are: ",even_login_hours)
print("Odd login hours are: ",odd_login_hours)

for hour in login_hours:
    if hour>5:
        peak_activity_login_hour.append(hour)
print("Peak activity Login hours are: ",peak_activity_login_hour)

login_hours.sort()
print("Login hours in ascending order: ",login_hours)
print("Count of peak hour logins: ",len(peak_activity_login_hour))
        
```
